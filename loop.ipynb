```python
l=[1,2,3,4,"sudh","kmumar",324,34.456,"abc"]
l1_num=[]
l2_str=[]
l3_float=[]
l4_bool=[]
for i in l:
  if type(i)==int:
    l1_num.append(i)
  elif  type(i)==float:
    l3_float.append(i)
  elif type(i)==bool:
    l4_bool.append(i)
  else:
    l2_str.append(i)
```
